```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
mpl.style.use('seaborn')
mpl.rcParams['font.family']='SimHei'#font.familiy,字体的名字,font.style字体风格,normal或italic,font.size,整数字号,或large,x-small等字符串
a=np.arange(10)
plt.plot(a,a*1.5,'r--',a,a*2,'mo',a,a*2.5,'cd')#','表示像素标记,极小点
plt.ylabel('纵坐标',fontproperties='SimHei',fontsize=20)
plt.axis([0,10,0,30])
plt.text(5,25,'$\mu=10$',fontsize=10)#相对坐标,相对坐标轴的坐标
plt.title('a',fontsize=25)
plt.annotate(r'$\mu=10$',xy=(4,10),xytext=(5,12),arrowprops=dict(facecolor='black',shrink=0.1,width=2))
#shrink表示起始到结束的缩进空白距离

plt.grid(True)
#plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use('seaborn')
#plt.pie?
labels=['a','b','c','d']
sizes=[15,30,45,10]
explode=(0,0,0,0.1)#设置离群距离
plt.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)
#plt.axis('equal')#正圆形
plt.show()
```

```python
import numpy as np
#plt.hist?

np.random.seed(10)
mu,sigma=100,20
a=np.random.normal(mu,sigma,size=100)

plt.hist(a,20,normed=1,histtype='bar',facecolor='g',alpha=0.5)
#bins=20,直方的个数,normed=1,归一化为频率,alpha控制透明度
plt.show()
```

```python

N=20#绘制的数据个数
theta=np.linspace(0,2*np.pi,N,endpoint=False)#将2pi等分成20分,不包含2pi
radii=10*np.random.rand(N)#生产10个0-10的随机数,作为长度
width=np.pi/4*np.random.rand(N)#生成10个(0,2.5pi)之间的随机数,作为宽度即角度

ax=plt.subplot(111,projection='polar')#设置极坐标绘图区
bars=ax.bar(theta,radii,width=width,bottom=0.0)#bottom起始坐标

for r,bar in zip(radii,bars):
    bar.set_facecolor(plt.cm.viridis(r/10))#Make a colormap from a list of colors
    bar.set_alpha(0.5)#设置不透明度
plt.show()
```
