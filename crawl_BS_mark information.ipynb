```python
#信息提取的方法
    #完整解析信息的标记形式,再提取关键信息,需要标记解析器,信息解析准确,但提取过程繁琐,速度慢
    #无视标记形式,直接搜索,只需对信息的文本查找函数,简单但结果可能不准确
    #融合方法,结合形式解析与搜索方法,提取关键信息,需要标记解析器及文本查找函数
    
#例子,提取HTML中所有URL链接
#1)搜索到所有<a>标签
#2)解析<a>标签格式,提取href后的链接内容
from bs4 import BeautifulSoup as BS
import requests as rq
import re 
url='http://finance.sina.com.cn/futures/quotes/OIL.shtml'
r=rq.get(url)
print(r.status_code)
print(r.encoding)
print(r.apparent_encoding)
r.encoding=r.apparent_encoding
demo=r.text
soup=BS(demo,'html.parser')
print(type(soup.find_all(['div','title'])))#搜索多个名称
#for link in soup.find_all('a'):
    
    #print(link.get('href'))
#for tag in soup.find_all(True):
    #print(tag.name)
for tag in soup.find_all(re.compile('^b')):#以含b的信息,作为查找要素
    print(tag.name)
#help(link.get)    
soup.find_all('a',attrs={'target':'_blank'},recursive=True)#不能同时搜索名,且搜索子节点的属性
print(type(soup.find_all('a',attrs={'target':'_blank'},recursive=True)))
print(soup.find_all('a',attrs={'target':'_blank'},recursive=True)[1].string)
print(type(soup.find_all('a',attrs={'target':'_blank'},recursive=True)[0]))
#for tag in soup.find_all('a',target="_blank"):
    #print(tag.name)
#soup.find_all(string=re.compile('黑色系'))
#soup.find_all(string=re.compile('华安宝利'))
print(soup.a(string='华安宝利'))
#soup(string=re.compile('华安宝利'))
#<tag>()等价于<tag>.find_all(...)
#soup()等价于soup.find_all(...)
```
