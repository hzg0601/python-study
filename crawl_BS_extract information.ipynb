```python
from bs4 import BeautifulSoup as BS#从bs4库中引用了一个类型BS
import requests as rq
url='https://python123.io/ws/demo.html'
r=rq.get(url)
demo=r.text
soup=BS(demo,'html.parser')#选择解析器
#html页面是以<>封装的信息
soup.prettify()
BS('<p>data</p>','html.parser')
#help(BS())
print(type(soup))
#HTML/XML库是,标签及其内容构成的标签树,bs4库是解析\遍历\维护\标签树的功能库
#<p class='title'>...</p>,p标签,<p,</p>名称Name成对出现, 而class='title'...称为标签Tag,class='title'为属性域,Attributes,0个或多个
#用于定义标签的特点,由键值对构成
sa=soup.find('a')
print(type(sa))
print(sa)
#BS库将标签树转换成BS类,HTML--标签树--BS类
soup.title
tag=soup.a#获取a标签
tag#如果有两个a标签,只捕获一个
soup.a.name
soup.a.parent.name
soup.a.parent.parent.name
```

```python
#BS类的基本元素
#Tag,标签,最基本的组织单元,分别<>和</>标明开头和结尾
    #Name,标签的名字,<p>...</p>的名字是'p',获取格式:<tag>.name
    #attibutes,标签的属性,字典形式组织,获取格式:<tag>.attrs
    
#NavigableString标签内非属性字符串,<>...</>中字符串,获取格式:<tag>.string,<tag>.text

#Comment,标签内字符串的注释部分,一种特殊的Comment类型

print(soup.title)
tag=soup.a#获取a标签
print(dir(tag))
print(tag.name,'tagname')
print(type(soup.a.name),'type_a_tagname')#如果有两个a标签,只捕获一个
soup.a.parent.name
print(soup.a.parent.parent.parent.name)
print(soup.a.attrs)#所有属性,字典类型
print(tag.attrs['href'])
print(type(tag.attrs['href']),'tag.attrs')
print(type(tag))
print(tag.string)
tagb=soup.p
print(tagb.string)#并不打印内层标签,navigablestring可以跨越标签层次
print(type(tagb.string))
newsoup=BS('<b><!--This is a comment--></b><p>This is not a comment</p>','html.parser')#comment的标识符,!--string--
print(newsoup.b.string)
print(type(newsoup.b.string))
print(newsoup.p.string)
print(type(newsoup.p.string))
```
