```python
from pyecharts import Map

city=['福建','天津','安徽','江西']
value=[2013,2016,1990,2008]
map=Map('迁徙','时间',width=800,height=600,title_color='#fff',title_pos='center',background_color='#404a59')

map.add('',city,value,maptype='china',is_label_show=True,is_visualmap=True,is_roam=False,
        visual_text_color='#fff',visual_range=[1990,2018],is_map_symbol_show=False,
       is_piecewise=True,visual_range_text=['',''],pieces=[{'max':2018,'min':2013,'label':'晚'},
                                                          {'max':2008,'min':1990,'label':'早'}])
#visual_text_color,图例文字颜色,is_map_symbol_show,是否显示省会/首都红点
#is_piecewise,是否分段,visual_range_text,分段标签,pieces,分段节点(可含标签)
map
```

```python
from pyecharts import Style,GeoLines
style=Style(title_color='white',title_pos='center',background_color='#404a59',width=900,height=600)

geolines=GeoLines('迁徙图','hzg',**style.init_style)
data=[['阜阳','抚州'],['抚州','厦门'],['厦门','天津']]

geolines.add('',data,is_legend_show=False,is_roam=False,geo_effect_symbol='arrow',is_geo_effect_show=True,
            geo_effect_color='purple',geo_effect_symbolsize=10,symbol_color='green',label_text_color='green',
             line_curve=0.3,line_opacity=0.5,label_color='yellow',is_label_show=True,label_formatter="{b}",symbol='pin'
            )
geolines.add('',[['黑河','保山']],line_curve=0,line_opacity=0.5,label_formatter='{b}',is_roam='False',symbol='pin',
            is_legend_show=False,is_geo_effect_show=False,line_width=2,line_color='red',is_label_show=True,
            label_text_color='green')
geolines
#label_formatter='{b}'才能以中文输出,否则以地理坐标输出
#line_curve=0,线条曲度
```

```python
from pyecharts import Polar,Grid,Page
from numpy import random
polar1=Polar('极坐标柱状角度图','hzg',height=600,width=900,title_pos='right')
index=['A','B','C','D','E','F','G']
value1=random.randint(1,7,7)
value2=random.randint(1,7,7)
value3=random.randint(1,7,7)
polar1.add('a',value1,radius_data=index,type='barAngle',is_stack=True)
polar1.add('b',value2,radius_data=index,type='barAngle',is_stack=True)
polar1.add('c',value3,radius_data=index,type='barAngle',is_stack=True)

polar2=Polar('极坐标柱状径图','hzg',height=600,width=900,title_pos='right')#不可修改图例位置

polar2.add('a',value1,radius_data=index,type='barRadius',is_stack=True)
polar2.add('b',value2,radius_data=index,type='barRadius',is_stack=True)
polar2.add('c',value3,radius_data=index,type='barRadius',is_stack=True)

grid=Grid()#第一个图需有x,y坐标轴
grid.add(polar2,grid_bottom='60%')
grid.add(polar1,grid_top='60%')
polar1

page=Page()
page.add(polar1)
page.add(polar2)
page
```

```python
from pyecharts import Map,Timeline

p=['安徽','江西','福建','天津']
value=[1990,2008,2013,2016]
map0=Map('迁徙','时间',width=800,height=600,title_color='#fff',title_pos='center',background_color='#404a59')

map0.add('',p,value,maptype='china',is_label_show=True,is_visualmap=True,is_roam=False,is_random=True,
        visual_text_color='#fff',visual_range=[1990,2018],is_map_symbol_show=False)
#visual_text_color,图例文字颜色,is_map_symbol_show,是否显示省会/首都红点
map3=map2=map1=map0
timeline=Timeline(timeline_bottom=0)
timeline.add(map0,value[0])
timeline.add(map1,value[1])
timeline.add(map2,value[2])
timeline.add(map3,value[3])
timeline
```
