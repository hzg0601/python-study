```python
#贪婪匹配和最小匹配
#re库默认采用贪婪匹配,即输出匹配最长的字符串

#最小匹配操作符,加问号
    #*?前一个字符的0次或无限次扩展,但输出最小匹配
    #+?前一个字符的1次或无限次扩展,但输出最小匹配
    #??前一个字符的0次或一次扩展,输出最小匹配
    #{m,n}扩展前一个字符m至n次,输出最小匹配
import re
match=re.search(r'PY.*N','PYANBNCNDN')
print(match.group(0))
match=re.search(r'PY.*?N','PYANBNCNDN')
print(match.group(0))
```
