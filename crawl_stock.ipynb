```python
#获取上交所和深交所的股票交易信息
#rq-bs4-re,仅能获取静态网页
import requests as rq
from bs4 import BeautifulSoup as BS
import re
import traceback as tc

def getHTMLText(url,code='utf-8'):#设置默认编码
    try:
        r=rq.get(url)
        r.raise_for_status()
        r.encoding=code
        return r.text
    except:
        return ''
def getStockList(lst,stockURL):
    html=getHTMLText(stockURL,'GB2312')#返回requests库的response对象
    soup=BS(html,'html.parser')#将response对象转为BS对象
    a=soup.find_all('a')#找到a标签
    for i in a:
        try:
            href=i.attrs['href']#取出属性href对应的值,字符串类型
            lst.append(re.findall(r'[s][hz]\d{6}',href)[0])
            ##re.findall,搜索字符串,以列表类型返回全部能匹配的子串,此处只取列表的第一个元素
        except:
            continue
        
def getStockInfo(lst,stockURL,fpath):
    count=0
    for stock in lst:
        url=stockURL+stock+'.html'
        html=getHTMLText(url)
        try:
            if html=='':
                continue
            infoDict={}
            soup=BS(html,'html.parser')
            stockInfo=soup.find('div',attrs={'class':'stock-bets'})#找出名称为div,属性class为stock-bets的标签
            
            name=stockInfo.find_all(attrs={'class':'bets-name'})[0]#找出所有属性class值为bets-name的标签,取出第一个
            
            infoDict.update({'股票名称':name.text.split()[0]})#取出name标签的string,并拆分它,取出第一个作为股票名称,后面还有数字
            
            keyList=stockInfo.find_all('dt')#允许,名称,属性,string的搜索
            valueList=stockInfo.find_all('dd')
            for i in range(len(keyList)):
                key=keyList[i].text
                val=valList[i].text
                infoDict[key]=val
            with open(fpath,'a',encoding='utf-8') as f:#a表示追加
                f.write(str(infoDict)+'\n')#写入文件,并退出循环
                count=count+1
                print('\r当前进度:{: .2f}%'.format(count*100/len(lst)),end='')#\r,将光标指向当前位置
        except:
            count=count+1
            print('\r当前进度:{: .2f}%'.format(count*100/len(lst)),end='')#%以控制百分比输出
            #tb.print_exc()
            continue
            

def main():
    stock_list_url='http://quote.eastmoney.com/stocklist.html'
    stock_info_url='https://gupiao.baidu.com/stock/'
    output_file='E:/BaiduStock.txt'
    slist=[]
    getStockList(slist,stock_list_url)
    getStockInfo(slist,stock_info_url,output_file)


main()
```
