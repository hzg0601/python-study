```python
import requests as rq
from bs4 import BeautifulSoup as BS
import bs4
def getHTMLText(url):
    #try:
    r=rq.get(url,timeout=30)
    r.raise_for_status
    print(r.status_code)
    r.encoding=r.apparent_encoding
    return r.text
    #except:
        #return ''
    
def fillUnivList(ulist,html):
    soup=BS(html,'lxml')#解析器,无法识别<td>1<td>
    for tr in soup.find('tbody').children:
        if isinstance(tr,bs4.element.Tag):
            tds=tr('td')#将所有标签名为td的标签存入数组
            #div=tds[1].find('div')
            ulist.append([tds[0].string,tds[1].string,tds[2].string,tds[3].string])
    pass

def printUnivList(ulist,num):
    tplt='{0:^10}\t{1:{4}^10}\t{2:^10}\t{3:^10}'#{4}表示当需要填充时用.format中index为4的字符填充
    print(tplt.format('排名','学校','省份','得分',chr(12288)))
    #print('{:>6}\t{:^15}\t{:^20}'.format('排名','学校','省份'))
    #^中间对齐,>右对齐,<左对齐,
    #可用数字表示顺序,
    ##1）不带编号，即“{}”
      ##2）带数字编号，可调换顺序，即{1}、{2}
      ##3）带关键字，即{a=}、{tom=}
        ##\t表示横向制表符,\v表示纵向制表符
        ##:10表示取10位
    #'b' - 二进制。将数字以2为基数进行输出。
    #'c' - 字符。在打印之前将整数转换成对应的Unicode字符串。
    #'d' - 十进制整数。将数字以10为基数进行输出。
    #'o' - 八进制。将数字以8为基数进行输出。
    #'x' - 十六进制。将数字以16为基数进行输出，9以上的位数用小写字母。
    #'e' - 幂符号。用科学计数法打印数字。用'e'表示幂。
    #'g' - 一般格式。将数值以fixed-point格式输出。当数值特别大的时候，用幂形式打印。
    #'n' - 数字。当值为整数时和'd'相同，值为浮点数时和'g'相同。不同的是它会根据区域设置插入数字分隔符。
    #'%' - 百分数。将数值乘以100然后以fixed-point('f')格式打印，值后面会有一个百分号
    for i in range(num):
        u=ulist[i]
        #print('{:>6}\t{:>15}\t{:>20}'.format(u[0],u[1],u[2]))
        print(tplt.format(u[0],u[1],u[2],u[3],chr(12288)))
#当中文字符宽度不够时,采用西文字符填充;中西文字符占用宽度不同
#chr(12288)表示中文空格


    
def main():
    uinfo=[]
    url='http://www.zuihaodaxue.cn/zuihaodaxuepaiming2017.html'
    html=getHTMLText(url)
    fillUnivList(uinfo,html)
    printUnivList(uinfo,20)
    
if __name__=='__main__':
    main()
    
```
